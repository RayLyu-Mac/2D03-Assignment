```python
y=[0,-8042,-11860,-13370,-13810,-13250,-12090,-10390,-8104,-5004,0]
import math
tang=[]
bs=[]
y1s=[]
dic={}
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(0,1,11)
display(x)
for i in range(len(x)):
    slope=112821*4*x[i]**3-263225*3*x[i]**2+249831*2*x[i]-99305
    b=y[i]-x[i]*slope
    tang.append(slope)
    bs.append(b)
    y1=slope+b
    y1s.append(y1)
fig=plt.figure()
ax=fig.add_subplot(1,3,1)
display(bs,y1s)
xt=np.linspace(-0.05,1,200)
for j in range(len(x)):
    yt=[]
    for jj in range(len(xt)):
        yt.append(xt[jj]*tang[j]+bs[j])
    ax.plot(xt,yt)
plt.xlabel('Compositon of Sn')
plt.ylabel('Gibbs Free Energy of Mixing')
plt.yticks(np.arange(0,-110000,-10000))
plt.grid(True)
ax.plot(x,y,linewidth=3, color='black')
plt.subplot(1,3,3),plt.plot(x,y)
plt.xlabel('Compositon of Sn')
plt.ylabel('Gibbs Free Energy of Mixing')
plt.yticks(np.arange(0,-15000,-1000))
plt.grid(True)
a1=[]
a2=[]
for kk in range(len(y1s)):
    a1.append(math.exp(y1s[kk]/1400/8.314))
    a2.append(math.exp(bs[kk]/1400/8.314))
dic['Xn for Sn']=x
dic['Gibbs Free Energy for Sn']=bs
dic['Activity for Sn']=a1
dic['Gibbs Free Energy for Cu']=y1s
dic['Activity for Cu']=a2
import pandas as pd
from pandas import DataFrame
df=DataFrame(dic)
display(df)
#plt.savefig('qqqq.png',dpi=1000)
```
